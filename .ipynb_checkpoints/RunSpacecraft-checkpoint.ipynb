## Dynamic model

Specify the values of the physical parameters:

```python
# Mass and MOI of base
mb = 6.
Jxb = 10.
Jyb = 10.
Jzb = 16.

# Mass and MOI of each wheel
mw = 1.
Jxw = 0.075
Jyw = 0.075
Jzw = 0.125
lw = 1.1
```

Derive the equations of motion:

```python
# Define yaw, pitch, roll yaw angles
psi, theta, phi = sym.symbols('psi, theta, phi')

# Define angular velocities
w_x, w_y, w_z = sym.symbols('w_x, w_y, w_z')

# Define torques
tau_1, tau_2, tau_3, tau_4 = sym.symbols('tau_1, tau_2, tau_3, tau_4')

# Compute resultant torques
lt = sym.nsimplify(lw) * sym.sqrt(2) / 2
T1 = - tau_1 * sym.Matrix([[lt], [0], [lt]])
T2 = - tau_2 * sym.Matrix([[-lt], [0], [lt]])
T3 = - tau_3 * sym.Matrix([[0], [lt], [lt]])
T4 = - tau_4 * sym.Matrix([[0], [-lt], [lt]])
T = T1 + T2 + T3 + T4

# Define MOI of spacecraft and wheels together
Jx = sym.nsimplify(Jxb + 4 * mw * lw**2)
Jy = sym.nsimplify(Jyb + 4 * mw * lw**2)
Jz = sym.nsimplify(Jzb + 4 * mw * lw**2)

# Define rotation matrices
Rz = sym.Matrix([[sym.cos(psi), -sym.sin(psi), 0], [sym.sin(psi), sym.cos(psi), 0], [0, 0, 1]])
Ry = sym.Matrix([[sym.cos(theta), 0, sym.sin(theta)], [0, 1, 0], [-sym.sin(theta), 0, sym.cos(theta)]])
Rx = sym.Matrix([[1, 0, 0], [0, sym.cos(phi), -sym.sin(phi)], [0, sym.sin(phi), sym.cos(phi)]])

# Define the transformation from angular velocity to angular rates
ex = sym.Matrix([[1], [0], [0]])
ey = sym.Matrix([[0], [1], [0]])
ez = sym.Matrix([[0], [0], [1]])
M = sym.simplify(sym.Matrix.hstack((Ry @ Rx).T @ ez, Rx.T @ ey, ex).inv(), full=True)

# Define euler's equations
euler = sym.Matrix([[(1 / Jx) * (T[0] + (Jy - Jz) * w_y * w_z)],
                   [(1 / Jy) * (T[1] + (Jz - Jx) * w_z * w_x)],
                   [(1 / Jz) * (T[2] + (Jx - Jy) * w_x * w_y)]])

# Define equations of motion
f = sym.simplify(sym.Matrix.vstack(M * sym.Matrix([[w_x], [w_y], [w_z]]), euler), full=True)
```

Function $f$ in EOM format:

$$\begin{bmatrix} \dot{\psi} \\ \dot{\theta} \\ \dot{\phi} \\ \dot{w_x} \\ \dot{w_y} \\ \dot{w_z} \end{bmatrix} = f\left(\psi, \theta, \phi, w_x, w_y, w_z, \tau_1, \tau_2, \tau_3, \tau_4\right)$$

Check function $f$'s format:

```python
f
```

```python
f_num = sym.lambdify((psi, theta, phi, w_x, w_y, w_z,tau_1, tau_2, tau_3, tau_4),f)
```

Set equilibrium values

```python
psi_e = 0
theta_e = 0
phi_e = 0
w_x_e = 0
w_y_e = 0
w_z_e = 0
tau_1_e = 0
tau_2_e = 0
tau_3_e = 0
tau_4_e = 0
```

Check set equilibrium values with function $f$

```python
print(f_num(psi_e, theta_e, phi_e, w_x_e, w_y_e, w_z_e, tau_1_e, tau_2_e, tau_3_e, tau_4_e))
```

Therefore, $f$ works with the set equilibrium conditions

Set up A and B matrices

```python
A_num = sym.lambdify([psi, theta, phi, w_x, w_y, w_z, tau_1, tau_2, tau_3, tau_4], f.jacobian([psi, theta, phi, w_x, w_y, w_z]))
A = A_num(psi_e, theta_e, phi_e, w_x_e, w_y_e, w_z_e, tau_1_e, tau_2_e, tau_3_e, tau_4_e)

B_num = sym.lambdify([psi, theta, phi, w_x, w_y, w_z, tau_1, tau_2, tau_3, tau_4], f.jacobian([tau_1, tau_2, tau_3, tau_4]))
B = B_num(psi_e, theta_e, phi_e, w_x_e, w_y_e, w_z_e, tau_1_e, tau_2_e, tau_3_e, tau_4_e)

print('A: ', A.tolist())
print('B: ', B.tolist())
```

Check controllability

```python
n = A.shape[0]

W = B

for i in range(1, n):
    col = np.linalg.matrix_power(A, i) @ B
    W = np.block([W, col])

r = np.linalg.matrix_rank(W)
print('W Matrix:')
print(W)
print('Rank:')
print(r)
print('Shape of A: ')
print(n)
```

The rank of the controllability matrix, $W$ is equal to the size of the matrix $A$. Therefore, the system is controllable.

## Sensor model

Set up symbolic variables for right ascension $\alpha$ and declination $\delta$ of each star:

```python
alpha, delta = sym.symbols('alpha, delta')
```

Specify the physical parameters:

```python
# Scope radius
r = 0.8 / 2.1
```

Derive the sensor model:

```python
# Position of star in space frame
p_star_in_space = sym.Matrix([[sym.cos(alpha) * sym.cos(delta)],
                              [sym.sin(alpha) * sym.cos(delta)],
                              [sym.sin(delta)]])

# Orientation of body frame in space frame
R_body_in_space = Rz * Ry * Rx

# Position of star in body frame (assuming origin of body and space frames are the same)
p_star_in_body = R_body_in_space.T * p_star_in_space

# Position of star in image frame
p_star_in_image = (1 / sym.nsimplify(r)) * sym.Matrix([[p_star_in_body[1] / p_star_in_body[0]],
                                                       [p_star_in_body[2] / p_star_in_body[0]]])

# Sensor model for each star
g = sym.simplify(p_star_in_image, full=True)
```

Set up $G$ matrix, which contains the positions of each star.

```python
for (i, star) in enumerate(simulator.stars):
    print(f'star {i} : alpha = {star["alpha"]:5.2f} : delta = {star["delta"]:5.2f}')

alpha0 = -0.10
delta0 = -0.15
alpha1 = 0.00
delta1 = -0.15
alpha2 = 0.10
delta2 = -0.15
alpha3 = 0.00
delta3 = 0.00
alpha4 = -0.10
delta4 = 0.15
alpha5 = 0.00
delta5 = 0.15
alpha6 = 0.10
delta6 = 0.15
```

```python
G = np.block([[g.subs([(alpha, alpha0), (delta, delta0)])],[g.subs([(alpha, alpha1), (delta, delta1)])],[g.subs([(alpha, alpha2), (delta, delta2)])],[g.subs([(alpha, alpha3), (delta, delta3)])],[g.subs([(alpha, alpha4), (delta, delta4)])], [g.subs([(alpha, alpha5), (delta, delta5)])], [g.subs([(alpha, alpha6), (delta, delta6)])]])
G = sym.Matrix(G)
G
```

Check $G$ with set equilibrium values

```python
G_num = sym.lambdify((psi, theta, phi), G)
Geq = G_num(psi_e, theta_e, phi_e)
Geq
```

Set up $C$ matrix

```python
C_num = sym.lambdify((psi, theta, phi), G.jacobian([psi, theta, phi, w_x, w_y, w_z]))
C = C_num(psi_e, theta_e, phi_e)
C
```

```python
Wo = C
for i in range(1,n):
    row = C @ np.linalg.matrix_power(A,i)
    Wo = np.block([[Wo],[row]])
rank = np.linalg.matrix_rank(Wo)
print(rank)
```

Define a controller that maps sensor measurements to actuator commands.

```python
class Controller:
    def __init__(self, K, L):
        self.dt = 0.01
        self.A = A
        self.B = B
        self.C = C
        self.K = K
        self.L =  L
        self.Geq = Geq.flatten()
        self.psi_e = 0
        self.theta_e = 0
        self.phi_e = 0
        self.w_x_e = 0
        self.w_y_e = 0
        self.w_z_e = 0
        self.tau_1_e = 0
        self.tau_2_e = 0
        self.tau_3_e = 0
        self.tau_4_e = 0
        
    
    def reset(self):
        self.xhat = np.zeros(6)
    
    def run(self, t, star_meas):
        """
        star_meas is a 1d array of length 14 - the image coordinates
        y_i and z_i of the i'th star (for i = 0, ..., 6) are at index
        2*i and 2*i+1 of this array, respectively
        """
        
        # FIXME: replace the following lines
        u = - self.K @ self.xhat
        front_torque = (u +self.tau_1_e)[0]
        back_torque = (u +self.tau_2_e)[1]
        left_torque = (u +self.tau_3_e)[2]
        right_torque = (u +self.tau_4_e)[3]
        
        y = star_meas-self.Geq.astype(float)
        self.xhat += self.dt * (self.A @ self.xhat + self.B @ u - self.L @ (self.C@self.xhat  - y))
        
        return front_torque, back_torque, left_torque, right_torque
```

Create function to obtain data from each simulation

```python
def get_data(K, L):
    
    # Call instance of controller
    controller = Controller(K, L)
    
    # Reset simulator with given initial conditions
    simulator.reset(
        orientation={
            'yaw': 0.,
            'pitch': 0.,
            'roll': 0.,
        },
        angular_velocity={
            'x': 0.,
            'y': 0.,
            'z': 0.,
        },
        scope_noise=0.1,)
    
    controller.reset()
    
    # Set data
    data = simulator.run(
    controller,           
    max_time=120.0,         
    data_filename=None,   
    video_filename=None,  
    print_debug=False,)    
    
    return data
```

Finally, including the dynamic and sensor model, gains matrices $K$ and $L$ need to be generated. Numerous $K$ matrices will be generated for testing purposes with the objective of minimizing state estimate errors.

#### Process for generating K matrices (from dynamic model)

```python
# Keep R constant. This will not be changed.
Re = np.eye(4)/5

'''
Input variables:
    end: Maximum limit for weights in q1 (yaw)
    end: Maximum limit for weights in q1 (yaw)
    end: Maximum limit for weights in q1 (yaw)
    end: Maximum limit for weights in q1 (yaw)
    end: Maximum limit for weights in q1 (yaw)
    end: Maximum limit for weights in q1 (yaw)
    num: Desired number of Q matrices to generate
    
Output variables:
    Q_list: List of Q matrices (size 6 x 6) 
'''

def generate_qmatrices(end_yaw, end_roll, end_pitch, end_wx, end_wy, end_wz, num):
    Q_list = []
    q1 = np.linspace(0.1, end_yaw, num)
    q2 = np.linspace(0.1, end_roll, num)
    q3 = np.linspace(0.1, end_pitch, num)
    q4 = np.linspace(0.1, end_wx, num)
    q5 = np.linspace(0.1, end_wy, num)
    q6 = np.linspace(0.1, end_wz, num)
    for i in range(0,len(q1)):
        Q = np.diag([q1[i], q2[i], q3[i],q4[i], q5[i], q6[i]])
        Q_list.append(Q)
    
    return Q_list

Q_list = generate_qmatrices(5, 5, 3, 1, 2, 2, 200)


'''
Input variables:
    Q_list: List of Q matrices (size 6 x 6) 
    Re: R matrix (size 4 x 4)
    
Output variables:
    K_list: List of K matrices
'''

def generate_dynamic_gains(Q_list, Re):
    K_list = []   # total K list
    P_list = []
    for i in range(0, len(Q_list)):
        P_list.append(linalg.solve_continuous_are(A, B, Q_list[i], Re))
        K_list.append(linalg.inv(Re) @ B.T @ P_list[i])
#         print(K_list)
            
    return K_list

K_list = generate_gains(Q_list, Re)[0]
```

Generate a list of $K$ matrices keeping a constant $L$ matrix

```python
# Set up observer Q and R matrices
Qo = np.eye(A.shape[0])
Ro = np.eye(C.shape[0])*0.4


P = linalg.solve_continuous_are(A.T, C.T, linalg.inv(Qo), linalg.inv(Ro)).T
# Sample L matrix generation for testing the K matrices
L = (linalg.inv(Ro) @  C @ P).T

data = []
times = []
yaws = []
pitches = []
rolls = []
    

working_K = []
working_Q = []
for i in range(len(K_list)):
    data.append(get_data(K_list[i], L))
    times.append(data[i]['t'])
    yaws.append(data[i]['yaw'])
    pitches.append(data[i]['pitch'])
    rolls.append(data[i]['roll'])
    if ((np.all(np.abs(yaws[i])) < 0.04) and (np.all(np.abs(pitches[i])) < 0.04) and (np.all(np.abs(rolls[i])) < 0.02) and (times[i][-1] > 45)) == True:
        if (times[i][-1] > 45) == True:
            working_K.append(K_list[i])
            working_Q.append(Q_list[i])
```
